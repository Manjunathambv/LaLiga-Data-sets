```python
# Create a new column with name “Winning Percent” and append it to the data set (10 points)
# Percentage of Winning = (GamesWon / GamesPlayed)*100
# If there are any numerical error, replace it with 0%
# Print the top 5 teams which has the highest Winning percentage
data1.GamesWon = data1.GamesWon.astype(int)
data1.GamesPlayed = data1.GamesPlayed.astype(int)

data1["winningPercent"] = (data1["GamesWon"] / data1["GamesPlayed"])*100
data1["winningPercent"]= data1["winningPercent"].replace(np.NaN, 0)
data1.head()

```

```python
# Group teams based on their “Best position” and print the sum of their points for all positions (20 points)
#         Eg: Best Position     Points
#          1                   25000
#          2                    7000

temp1 = data1.groupby(by = ["BestPosition"])
groupeddata = temp1[["Points"]].sum().add_prefix("sumOf_").reset_index()
groupeddata
```
